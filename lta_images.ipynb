```python
import numpy as np
import pandas as pd
import requests, re, zipfile, logging, sys
from datetime import datetime
from pathlib import Path

import folium
from IPython.display import display
```

```python
traffic_images_url = "http://datamall2.mytransport.sg/ltaodataservice/Traffic-Imagesv2"
api_key = "AO4qMbK3S7CWKSlplQZqlA=="

headers = {
    'AccountKey' : api_key
}
```

```python
# Load the camera info CSV file
df = pd.read_csv('camera_info.csv')

# Initialize a map centered on Singapore
singapore_map = folium.Map(location=[1.3521, 103.8198], zoom_start=12)

# Iterate over the rows of the DataFrame and add points to the map
for index, row in df.iterrows():
    lat = row['Latitude']
    lon = row['Longitude']
    camera_id = int(row['CameraID'])
    
    # Add a marker with a popup showing the camera ID and a tooltip for mouseover
    folium.Marker(
        location=[lat, lon],
        popup=f'Camera ID: {camera_id}',
        tooltip=camera_id  # Show the camera ID on hover
    ).add_to(singapore_map)

# Display the map in the notebook
display(singapore_map)
```

```python
outzipfile = "lta_images.zip"

try:
    r3 = requests.get(traffic_images_url, headers=headers)

    # retrieve images
    for ii in r3.json()['value']:
        img_link = ii['ImageLink'] 
        xx = re.search('.*(?<=.jpg)', img_link) 
        fname = re.split('/', xx.group())[-1]
        r4 = requests.get(img_link, headers=headers)
        
        with open(fname, 'wb') as f:
            f.write(r4.content)
            print(f'{fname} written..')
      
        with zipfile.ZipFile(outzipfile, 'a', zipfile.ZIP_DEFLATED) as ff:
            ff.write(fname)
            print(f'{fname} written to zip..')

        Path(fname).unlink()
    
except Exception as err: 
    print(f"Error {err} occurred.") 
```
